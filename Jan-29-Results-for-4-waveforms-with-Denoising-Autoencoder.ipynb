# This page shows the reconstruction of CCSN waveforms using an autoencoder

## 1. Autoencoder
### The autoencodoer consists of two networks: a convolutional neural network as as encoder and a two-layer long-short term memory decoder.
### The autoencoder was trained to denoise input whitened time series


## 2. Data generation:

### 1. 2 magnetorational waveforms from Richers, et al 2017 were used.
### 2. 2 neutrino waveforms from Ott, et al 2013 were used.
### 3. These original waveforms were first passed through a high pass filter to filter out low frequency component (< 11 Hz)
### 4. A tukey window with alpha = 0.08 was then applied to every waveform.
### 5. The waveform with the longest length was then identified, each of the rest waveforms was zero padded to this length, and scaled to optimal SNR = 12. 
### 6. 6000 noise realisations of length 1.7 times longer than the duration of the waveforms were generated, using the power spectral density of aLIGO at design sensitivity.
### 8. The waveforms were then duplicated 1000 times making them 4000 in total.
### 9. 4000 sky locations were generated randomly. Each of the 4000 waveforms was added to a different noise realisation, with the start of the waveform in the noise realisation being randomised.
### 10. After step 9, there were 6000 time series: 4000 (noise realisations + signals) and 2000 (noise realisations).
### 11. Each time series was then padded with 64 zeros at the beginning and 64 zeros at the end.
### 12. These time series were then whitened, split in segments so that the first segment contain the first 129 time stamps of the first time series, and the second segment contained the 2nd to 130th time stamps of the first waveform, and so on for all time series.
### 13. After step 12, there will be X segments, where X = the number of time series multiplied by the number of time stamps of each time series. Each segment contained 129 values.
### 14. The segments in step 13 were then used as training/validation/test data, i.e. 80% of the segments used for training, 10% for validation, and 10% for testing.
### 15. The target data were the noise free values at the middle of each segment.
### 16. For example, when inputting the first segment (1st time stamp to 129 th time stamp, the autoencoder was tasked to predict the noise free value of the 65th time stamp). 
### 17. To quantify the similarity between the denoised time series and the noise-free signal, a quanity defined below was used,
